```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data = {'Product_Category': ['A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C'],
        'Rating': [5, 4, 3, 4, 5, 2, 3, 1, 5, 4]}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(6, 6))
sns.countplot(x='Rating', hue='Product_Category', data=df, palette='viridis')
plt.title('Distribution of Ratings for Different Product Categories')
plt.xlabel('Rating')
plt.ylabel('Count')
```

```python
# Sample dataset
data = {'Brand': ['A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C'],
        'Price': [100, 150, 200, 180, 220, 90, 120, 130, 110, 140]}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(6, 6))

# Create a boxplot
sns.boxplot(x='Brand', y='Price', data=df, palette='Set3')
plt.title('Distribution of Prices Across Different Brands')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.show()
```

```python
# Sample dataset
data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05'],
        'Review_Score': [4.5, 3.8, 4.2, 4.0, 4.6]}
df = pd.DataFrame(data)

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and year for grouping
df['Month_Year'] = df['Date'].dt.to_period('M').astype(str)  # Convert to string

# Calculate average review score per month
avg_review_score = df.groupby('Month_Year')['Review_Score'].mean().reset_index()

# Plot
plt.figure(figsize=(6, 6))
sns.lineplot(x='Month_Year', y='Review_Score', data=avg_review_score, marker='o')
plt.title('Average Review Score Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Average Review Score')
plt.xticks(rotation=45)
plt.grid(axis='x')
```
